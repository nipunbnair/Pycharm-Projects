```python
import itertools
ParentMass = 484
Amino="GASPVTCILNDKQEMHFRYW"
AminoWeight={'G':57,'A':71,'S':87,'P':97,'V':99,'T':101,'C':103,'I':113,'L':113,'N':114,'D':115,'K':128,'Q':128,'E':129,'M':131,'H':137,'F':147,'R':156,'Y':163,'W':186}
def getAminoWeight(x):
    w=0
    for i in x:
        w=w+AminoWeight[i]
        
    return w
```

```python
def permutate_all_substrings(text):
    permutations = []
    for length in range(1,4):
        for p in itertools.permutations(text, length):
              permutations.append("".join(p))
    return permutations
        
AllPossible = permutate_all_substrings(Amino)
print(len(AllPossible))
AllPossible
```

```python
possible=[]
for i in AllPossible:
    if ParentMass==getAminoWeight(i):
        print(i,getAminoWeight(i))
        possible.append(i)
```

```python
#2
def generateTheoreticalSpectrum(string):
    temp=permutate_all_substrings(string)
    Spectrum=[]
    for i in temp:
        Spectrum.append([getAminoWeight(i),i])
        
    return Spectrum

TempResult=generateTheoreticalSpectrum("NQEL")

LinearSpectrum=[[0,""]]+TempResult
LinearSpectrum
```
