### Q1)Write a program to generate a random genome sequence of length 15.

```python
import timeit

from random import choice
def dnaseq(length):
    rand_DNA=""
    for count in range(length):
        rand_DNA+=choice("GTAC")
    return rand_DNA
```

```python
d=dnaseq(15) 
print(d)
```

### Q2)Write a program to generate RNA sequence from the DNA sequence generated from the above genome sequence.

```python
def DNAToRNA(dnaseq):
    RNASeq = ''
    for i in dnaseq:
        if i == 'A':
            RNASeq += 'U'
        elif i == 'C':
            RNASeq += 'G'
        elif i == 'T':
            RNASeq += 'A'
        else:
            RNASeq += 'C'
    return RNASeq
```

```python
y=DNAToRNA(d)
print('RNA Sequence is :',y)
```
