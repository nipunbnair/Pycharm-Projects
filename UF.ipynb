```python
class UF:
    def __init__(self,n):
        self.id=[]
        self.N= n
        for i in range(0,n):
            self.id.append(i)
            
    def connected(self,p,q):
        if self.id[p]==self.id[q]:
            return True
        else: 
            return False
        
    def union(self,p,q):
        if not(self.id[p]==self.id[q]):
            p1=self.id[p]
            q1=self.id[q]
            for i in range(self.N):
                if (self.id[i]==p1):
                    self.id[i]=q1
                    
    def display(self):
        for i in range(self.N):
            print(self.id[i]," ")
           
        
        
G1=UF(10)
G1.display()
print(" ")
G1.union(2,3)
G1.display()
print(" ")
G1.union(1,3)
G1.display()
print(" ")
G1.union(5,8)
G1.display()
print(" ")
G1.union(2,6)
G1.display()
print(" ")
G1.union(1,4)
G1.display()
print(" ")
G1.display()
# Quick find
print(G1.connected(2,4))  
print(G1.connected(5,8))
```
